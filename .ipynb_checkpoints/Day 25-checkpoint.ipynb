```python
from collections import defaultdict

class TurningMachine:
    def __init__(self, steps=6):
        self._tape = defaultdict(int)
        self._cursor = 0
        self._run_machine(steps)
        
    @property
    def checksum(self):
        return sum(self._tape.values())
        
    def _run_machine(self, steps):
        state = self._state_A
        
        for _ in range(steps):
            state = state()
    
    def _state_A(self):
        current_value = self._tape[self._cursor]
        
        if current_value == 0:
            self._tape[self._cursor] = 1
            self._cursor_right()
            new_state = self._state_B
        elif current_value == 1:
            self._tape[self._cursor] = 0
            self._cursor_left()
            new_state = self._state_C
        else:
            raise NotImplementedError()
            
        return new_state
    
    def _state_B(self):
        current_value = self._tape[self._cursor]
        
        if current_value == 0:
            self._tape[self._cursor] = 1
            self._cursor_left()
            new_state = self._state_A
        elif current_value == 1:
            self._tape[self._cursor] = 1
            self._cursor_left()
            new_state = self._state_D
        else:
            raise NotImplementedError()
            
        return new_state
    
    def _state_C(self):
        current_value = self._tape[self._cursor]
        
        if current_value == 0:
            self._tape[self._cursor] = 1
            self._cursor_right()
            new_state = self._state_D
        elif current_value == 1:
            self._tape[self._cursor] = 0
            self._cursor_right()
            new_state = self._state_C
        else:
            raise NotImplementedError()
            
        return new_state
    
    def _state_D(self):
        current_value = self._tape[self._cursor]
        
        if current_value == 0:
            self._tape[self._cursor] = 0
            self._cursor_left()
            new_state = self._state_B
        elif current_value == 1:
            self._tape[self._cursor] = 0
            self._cursor_right()
            new_state = self._state_E
        else:
            raise NotImplementedError()
            
        return new_state
            
    def _state_E(self):
        current_value = self._tape[self._cursor]
        
        if current_value == 0:
            self._tape[self._cursor] = 1
            self._cursor_right()
            new_state = self._state_C
        elif current_value == 1:
            self._tape[self._cursor] = 1
            self._cursor_left()
            new_state = self._state_F
        else:
            raise NotImplementedError()
            
        return new_state
    
    def _state_F(self):
        current_value = self._tape[self._cursor]
        
        if current_value == 0:
            self._tape[self._cursor] = 1
            self._cursor_left()
            new_state = self._state_E
        elif current_value == 1:
            self._tape[self._cursor] = 1
            self._cursor_right()
            new_state = self._state_A
        else:
            raise NotImplementedError()
            
        return new_state
    
    def _cursor_right(self):
        self._cursor += 1
        
    def _cursor_left(self):
        self._cursor -= 1
```

```python
%%time
steps = 12656374
print(TurningMachine(steps=steps).checksum)
```
