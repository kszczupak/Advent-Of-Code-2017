```python
class ExpressionParser:
    def __init__(self, expressions):
        self._values_dictionary = dict()
        self.max_peek = 0
        self._process_expressions(expressions)
        
    def get_max_value(self):
        return max(self._values_dictionary.values())
    
    def _process_expressions(self, expressions):
        for expression in expressions.splitlines():
            action, action_variable, condition = self._parse_expression(expression)

            if condition:
                action(action_variable)
        
    def _parse_expression(self, expression):
        (action_variable, action_function, action_value, _,
        condition_variable, condition_function, condition_value) = expression.split(" ")

        action = None
        condition = None

        self._create_if_not_exist(action_variable)
        self._create_if_not_exist(condition_variable)
        action_value = int(action_value)
        
        def increment(value_name):
            self._values_dictionary[value_name] += action_value
            self._process_peek(value_name)
            
        def decrement(value_name):
            self._values_dictionary[value_name] -= action_value
            self._process_peek(value_name)

        if (action_function == "inc"):
            action = increment
        elif (action_function == "dec"):
            action = decrement
        else:
            raise NotImplementedError()

        condition = eval(f"self._values_dictionary['{condition_variable}'] {condition_function} {condition_value}")

        return action, action_variable, condition

    def _create_if_not_exist(self, variable_name):
        try:
            self._values_dictionary[variable_name]
        except KeyError:
            self._values_dictionary[variable_name] = 0
    
    def _process_peek(self, value_name):
        if self._values_dictionary[value_name] > self.max_peek:
            self.max_peek = self._values_dictionary[value_name]
```
